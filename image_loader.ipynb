```python
from dotenv import load_dotenv
import os, pandas as pd
import requests
import logging
load_dotenv()
api_key = '441cc11115703944207561d5e14c8a76'

import re

links = pd.read_csv("data/ml-32m/links.csv")
def get_tmdbId (links_df , movieId):
    id = links_df[links_df['movieId']==movieId]['tmdbId']
    return int(id.iloc[0])

movies = pd.read_csv("data/ml-32m/movies.csv")
```

```python
print(api_key)
```

```python
for t in movies["movieId"].head(10):
    url = f"https://api.themoviedb.org/3/movie/{get_tmdbId(links,t)}"
    params = {"api_key": api_key}
    res = requests.get(url, params=params).json()
    q = res.get("poster_path")
    print (f"https://image.tmdb.org/t/p/w500{q}" if q else None)
```
